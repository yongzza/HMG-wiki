# mtcars 데이터셋 분석

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('mtcars.csv')
data
```

```python
# mpg = 갤런 당 주행 거리 (마일), 자동차의 연비를 나타내는 지표
# cyl = 실린더 수, 엔진의 성능과 효율에 영향
# disp = 배기량, 클수록 더 많은 공기와 연료를 연소
# hp = 최대 마력, 엔진의 출력 성능을 나타내는 지표
# drat = 후방 차축 비율, 차량의 가속 성능과 연비에 영향
# wt = 차량 무게
# qsec = ¼ 마일 시간, 가속 성능을 측정하는 지표
# vs = 엔진의 배치 방식 (0=V형, 1=직렬)
# am = 변속기 (0 = 자동, 1 = 수동)
# gear = 전진 기어 수, 차량의 주행 성능과 연비에 영향
# carb = 카퓨레터 수, 연료와 공기를 혼합하는 장치
```

```python
# gear, transmission features에는 각각 몇가지 값이 있는지를 출력하세요.
data['gear'].nunique()
```

```python
data['am'].nunique()
```

```python
# gear와 transmission의 경우의 수 조합별로 나누고 각 조합에 몇개의 차량이 있는지를 알아내세요.
# 1)pandas.crosstab method를 사용하는 방법
result1 = pd.crosstab(data['gear'], data['am'])
result1
```

```python
# 바 그래프로 그리세요
result1.plot(kind='bar', figsize=(10, 6))
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')
plt.xticks(rotation=0)
plt.show()
```

```python
# 2) 사용하지 않는 방법
# .size(): 그룹별로 행의 수 반환
# reset_index를 하지 않으면 멀티인덱스 시리즈 형태
result2 = data.groupby(['gear', 'am']).size().reset_index(name='cnt')
result2
```

```python
# 결과를 바 그래프로 시각화
# .fillna(0): 결측값으로 인해 시각화 문제 방지
result2 = result2.pivot(index='gear', columns='am', values='cnt').fillna(0)

result2.plot(kind='bar', figsize=(10, 6))
plt.title('# of Cars by Gear x Transmission')
plt.xlabel('(Gear, Auto Transmission)')
plt.ylabel('# of Cars')
plt.xticks(rotation=0)
plt.show()
```

## Cylinder와 HorsePower 간의 상관 관계

```python
# 이런 데이터셋을 분석해서 얻을 수 있는 경제적 가치는 무엇일까요? 어떤 비즈니스 상황에서 이런 데이터셋을 분석해서 어떤 경제적 가치를 얻을 수 있을까요?

- 차량 판매 시, 고객이 중시하는 성능 지표(mpg: 자동차의 연비, disp: 배기량)를 반영한 마케팅 전략을 수립하여 고객 맞춤 차량을 판매할 수 있음
- 엔지니어가 차량을 설계할 때, cyl(실린더 수), gear 등 차량의 성능에 영향을 미치는 것을 높이면서 wt(차량 무게)는 줄이는 효율적인 차량 설계를 할 수 있음
```

```python
# 변수들 간의 상관 관계가 높은 조합을 임의로 2개 선택해서 해당 데이터 간의 상관 관계를 그래프로 그리고 어떤 결론을 내릴 수 있는지를 토의하세요.

# cyl, disp -> 0.9
# disp, wt -> 0.89
# mpg, wt -> -0.87
# mpg, cyl -> -0.85
 
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['cyl'], data['disp'], alpha=0.7)
plt.xlabel('cyl')
plt.ylabel('disp')
plt.show()
```

```python
# 실린더 수 증가 -> 엔진의 크기와 배기량 커짐
# 고객이 연비를 중시할 경우, 적은 실린더 수와 적당한 배기량을 가진 효율적인 엔진 설계를 할 수 있음
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['disp'], data['wt'], alpha=0.7)
plt.xlabel('disp')
plt.ylabel('wt')
plt.show()
```

```python
# 배기량이 커질수록 무게가 증가하는 경향이 있음(양의 상관관계) 
# 차량이 무거울수록 더 많은 연료를 소비하므로 고성능차를 설계할 때 연비를 고려해야 함
```
